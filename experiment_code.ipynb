```python
from langchain import HuggingFaceHub, LLMChain
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from langchain.prompts import PromptTemplate
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.agents import initialize_agent, Tool
from langchain.agents.agent_types import AgentType
from langchain.tools import tool
import os
from dotenv import load_dotenv

# Load environment variables (HuggingFace API key)
load_dotenv()
os.environ["HUGGINGFACEHUB_API_TOKEN"] = os.getenv("HUGGINGFACEHUB_API_TOKEN")

# 1. Set Up HuggingFace LLM
llm = HuggingFaceHub(
    repo_id="meta-llama/Llama-3.2-3B-Instruct",  
    model_kwargs={"temperature": 0.7, "max_length": 512}
)
```

```python


# 2. Define the Prompt Template
prompt_template = """
You are a helpful RAG chatbot assisting the user by answering their queries.
Use the following context to answer the user's question concisely and accurately:
if user not ask about SR University tell him "I specially made for SR univerity, i cat provide other informations"

Context: {context}
Question: {question}

Helpful Answer:"""

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=prompt_template
)

# 3. Memory for Conversation
memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)


retrieval_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=vectordb.as_retriever(search_kwargs={"k": 5}),
    memory=memory,
    combine_docs_chain_kwargs={"prompt": prompt}
)

# Register the Tool
rag_tool = Tool(
    name="RAG Retrieval",  # Tool name used in the agent
    func=retrieval_chain.run,
    description="Use this tool to retrieve relevant context and answer user queries."
)

# Initialize the Agent
from langchain.agents import AgentType

tools = [rag_tool]
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Run the Agent
print("RAG Chatbot Initialized! Type 'exit' to stop.")

while True:
    user_input = input("\nYou: ")
    if user_input.lower() == "exit":
        break
    response = agent.run(user_input)
    print(f"Agent: {response}")
```
